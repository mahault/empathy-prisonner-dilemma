We use a `SearchAgent` in a graph environment, with a single agent.

## Simulation 1 - find object, no prior

Agent has uniform prior beliefs over object location.

```python
import util
from envs import GraphEnv
from agents import SearchAgent
from experiment import tick, stringify
from visualize import *

# generate Environment
graph, metadata = util.generate_connected_clusters(3, 4)
agent_config = {
    "Agent 1": 4,
}
object_config = {"Reward": 6}
env = GraphEnv(graph, agent_config, object_config)

# generate a single Agent
agent = SearchAgent(list(agent_config.keys())[0], env, objects_of_interest=[0])


result = {a: [] for a in agent_config.keys()}
T = 10
for t in range(T):
    info = tick(agent, env)
    for a, info in info.items():
        result[a].append(info)

plot_beliefs(result)
plot_circumplex(result)
```

# Simulation 2 - find object, correct prior

We give the agent a correct prior belief on where the object is.

```python
import util
from envs import GraphEnv
from agents import SearchAgent
from experiment import tick, stringify
from visualize import *

# generate Environment
graph, metadata = util.generate_connected_clusters(3, 4)
agent_config = {
    "Agent 1": 1,
}
object_config = {"Reward": 6}
env = GraphEnv(graph, agent_config, object_config)

# generate a single Agent
agent = SearchAgent(list(agent_config.keys())[0], env, objects_of_interest=[0])

# set prior
D = agent.D
D[1][6] = 10.0
agent.set_prior(D)

result = {a: [] for a in agent_config.keys()}
T = 10
for t in range(T):
    info = tick(agent, env)
    for a, info in info.items():
        result[a].append(info)

plot_beliefs(result)
plot_circumplex(result)
```

# Simulation 5 - uniform prior, but there is no object

The generative model of the agent states that the object _must_ be at some location. The agent keeps going around in circles trying to find the object.

```python
import util
from envs import GraphEnv
from agents import SearchAgent
from experiment import tick, stringify
from visualize import *

# generate Environment
graph, metadata = util.generate_connected_clusters(3, 4)
agent_config = {
    "Agent 1": 4,
}
object_config = {"Reward": -1}
env = GraphEnv(graph, agent_config, object_config)

# generate a single Agent
agent = SearchAgent(list(agent_config.keys())[0], env, objects_of_interest=[0], not_here=False)


result = {a: [] for a in agent_config.keys()}
T = 50
for t in range(T):
    info = tick(agent, env)
    for a, info in info.items():
        result[a].append(info)

plot_beliefs(result)
plot_circumplex(result)
```
